# Data Structures (List, Set, Dictionary)

Difference between Data Structures:

1. Tuple: ()
*   It can hold any data type
*   It is an ordered data structure (It contains "Index" function)
*   Tuples are immutable(Unchangable)
*   Eg: Dataset of Unique Values such as

2. List:[]
*   It can hold any data type
*   It is an ordered data structure (It contains "Index" function)
*   Lists are mutable(Changable)
*   Eg: Dataset of Common Values such as Name, Address, etc.,

3. Set: {}
*   It can hold any data type
*   It is unordered data structure
*   Sets are mutable(Changable)
*   Eg: Dataset of unique values such as Pan Number, Aadhar Number, etc.,

4. Dictionary: {key : value}
*   It can hold any data type
*   It is unordered data structure
*   Dictionary is mutable
*   Eg: {key : value} pair => {'name' : 'syed'} {'age' : 25}

**Functions available in List Data Structure:**

**Append** - Adding an element in the dataset

**Clear** - Removing all the elements in the dataset

**Copy** - Helps in copying the dataset and storing it in another value

**Count** - Counting the no.of given element inside the dataset

**Extend** - Extends a dataset by  elements from another dataset

**Index** - To check the position of the element in the dataset (First Occurence)

**Insert** -Adding an element in the desired position

**Pop** - Removes the last element or the element in the given position

**Remove** - Removes the First Occurence of the given element

**Reverse** - Helps in reversing the order of elements in a dataset

**Sort** - Arranges the elements in the dataset (Ascending order)

**Functions available in Set Data Structure:**

**add** - Adds an element to the dataset

**clear** - Clears the dataset completely

**copy** - Copies the  dataset completely

**difference** - Gives elements that are not common in both sets

**difference_update** - Removes same elements from the dataset and returns the updated dataset

**discard** - Removes the mentioned element in the dataset

**intersection** - Gives the common elements in both sets

**intersection_update** - Removes different elements from the dataset and returns the updated dataset

**isdisjoint** - Will check both datasets for common elements and returns True or False

**issubset** - Will check if the dataset-1 is a subset of dataset-2 and returns True or False

**issuperset** - Will check if the dataset-1 is the original set of dataset-2 and returns True or False

**pop** - Removes an random element in the dataset

**remove** - Removes the mentioned element in the dataset

**symmetric_difference** - Will remove similar elements from both datasets and gives the distinct elements

**symmetric_difference_update** - Will remove similar elements from both datasets and updates the original dataset with distinct elements

**union** - Will join all the elements in both the datasets and returns without duplicate elements

**update** - Will add distinct elements of set1 to set2

**Functions available in Dictionary Data Structure:**

**clear** - Clears the dataset completely

**copy** - Copies the dataset compeletely

**fromkeys** - Creates new dictionary with specified keys and values

**get** - Returns the value of the specified key

**items** - Returns the list of key, value pair in the whole dictionary

**keys** - Returns the list of key in the whole dictionary

**pop** - Removes the mentioned key, value pair

**popitem** - Removes the last key, value pair

**setdefault** - Returns the value associated with the specified key. If the key does not exist, it inserts the key with the specified value and returns the value.

**update** - Updates the dictionary with elements from another dictionary

**values** - Returns the list of values in the whole dictionary

Dictionary {key:value}

*   Key names should be unique
*   Values can be same
