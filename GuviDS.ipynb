# Data Structures (List, Set, Dictionary)

Difference between Data Structures:

1. Tuple: ()
*   It can hold any data type
*   It is an ordered data structure (It contains "Index" function)
*   Tuples are immutable(Unchangable)
*   Eg: Dataset of Unique Values such as

2. List:[]
*   It can hold any data type
*   It is an ordered data structure (It contains "Index" function)
*   Lists are mutable(Changable)
*   Eg: Dataset of Common Values such as Name, Address, etc.,

3. Set: {}
*   It can hold any data type
*   It is unordered data structure
*   Sets are mutable(Changable)
*   Eg: Dataset of unique values such as Pan Number, Aadhar Number, etc.,

4. Dictionary: {key : value}
*   It can hold any data type
*   It is unordered data structure
*   Dictionary is mutable
*   Eg: {key : value} pair => {'name' : 'syed'} {'age' : 25}

```python
# Nested for loop (loop inside a loop)
a=['apple','orange','banana']
for i in a:
  print(i)
for j in i:
  print(j)
for k in j:
  print(k)
```

```python
a=['apple','orange','banana']
for i in a:
  print(i)
  for j in i:
    print(j)
    for k in j:
      print(k)
```

The above mentioned code features three nested for loops. The outermost loop iterates through the elements of the list a, which contains the strings 'apple', 'orange', and 'banana'. For each string, the first inner loop then iterates through every character of that string. Finally, the innermost loop iterates through each character obtained from the second loop. Since j in for k in j will always be a single character, the innermost loop effectively prints each character twice. For example, when i is 'apple', it will first print 'apple', then for 'a', it will print 'a' twice, then for 'p', it will print 'p' twice, and so on for all characters in 'apple', 'orange', and 'banana'.
