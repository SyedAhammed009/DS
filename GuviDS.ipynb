Difference between Data Structures:

1. Tuple:
*   It can hold any data type
*   It is an ordered data structure (It contains "Index" function)
*   Tuples are immutable(Unchangable)
*   Eg: Dataset of Unique Values such as PanCard Number, Aadhar Card number

2. List:
*   It can hold any data type
*   It is an ordered data structure (It contains "Index" function)
*   Tuples are mutable(Changable)
*   Eg: Dataset of Common Values such as Email, Phone Number


```python
dir(list) # Shows the built-in Functions of List Data Type
# The functions with underscore are fixed and cannot be changed, function without underscore can be used to perform operations
```

**Functions available in List Data Structure:**

Append - Adding an element in the dataset

Clear - Removing all the elements in the dataset

Copy - Helps in copying the dataset and storing it in another value

Count - Counting the no.of given element inside the dataset

Extend - Extends a dataset by  elements from another dataset

Index - To check the position of the element in the dataset (First Occurence)

Insert -Adding an element in the desired position

Pop - Removes the last element or the element in the given position

Remove - Removes the First Occurence of the given element

Reverse - Helps in reversing the order of elements in a dataset

Sort - Arranges the elements in the dataset (Ascending order)

```python
a = [1,2,3,4,5,6,1]
a.append(7)
print(a)
```

```python
a1 = a.copy()
print(a1)
```

```python
a1.clear()
print(a1)
```

```python
a.count(1)
```

```python
b=[8,9,10]
a.extend(b)
print(a)
```

```python
a.index(1)
```

```python
a.insert(4,5) #(position of the element, Value of the element)
print(a)
```

```python
a.pop()
print(a)
```

```python
a.pop(4)
print(a)
```

```python
a.remove(1)
print(a)
```

```python
a.reverse()
print(a)
```

```python
a.sort()
print(a)
```
